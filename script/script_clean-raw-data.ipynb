```python
import pandas as pd
```

```python
data = pd.read_csv("../data/number-of-deaths-by-risk-factor.csv")
```

```python
data_cleaned = data.drop(['High cholesterol (deaths)'],axis=1)
data_cleaned = data_cleaned.dropna()
data_cleaned = data_cleaned.rename(columns = {data_cleaned.columns[0]: 'Country'})
data_cleaned.columns = data_cleaned.columns.str.strip().\
    str.lower().\
    str.replace(' ', '_').\
    str.replace(' ', '_'). \
    str.replace('(', '').str.replace(')', ''). \
    str.replace(r'_deaths','')
```
